```python
neighborhoods = pd.read_csv("neighborhoods.csv",
index_col=[0, 1, 2])
neighborhoods
```

```python
neighborhoods = pd.read_csv("neighborhoods.csv",
index_col=[0, 1, 2],
header=[0, 1])
neighborhoods
```

## Sort MultiIndex DataFrame

```python
neighborhoods.sort_index(level=2)
```

```python
neighborhoods.sort_index(level=[1,2], ascending=[True, False])
```

```python
neighborhoods.sort_index(axis=1, level="Category", ascending=False)
```

```python
neighborhoods.sort_index(axis=1, level="Subcategory", ascending=True)
```
