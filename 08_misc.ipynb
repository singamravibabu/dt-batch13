# RESHAPING AND PIVOTING

```python
import pandas as pd
```

```python
sales = pd.read_csv("sales_by_employee.csv", parse_dates=["Date"])
```

```python
sales
```

```python
sales.pivot_table(
    index="Date",
    columns="Name",
    values="Revenue",
    aggfunc="count",
    fill_value=0,
    margins=True,
    margins_name="Total"
)
```

```python
sales_by_date_name = sales.pivot_table(
    index="Name",
    columns="Date",
    values="Revenue",
    aggfunc="sum"
)
sales_by_date_name
```

```python
sales_by_date_name.stack()
```

```python
sales_by_name_customer = sales.pivot_table(
    index=["Name", "Customer"],
    values="Revenue",
    aggfunc="max"
)
sales_by_name_customer
```

```python
sales_by_name_customer.unstack()
```

## MELTING DATA SETS

```python
video_game_sales = pd.read_csv("video_game_sales.csv")
video_game_sales
```

```python
video_game_sales.melt(id_vars="Name",
 value_vars="NA",
 var_name="Continents",
 value_name="Sales")
```

```python
video_game_sales.melt(id_vars="Name",
 value_vars=["NA", "EU", "JP", "Other"],
 var_name="Continents",
 value_name="Sales")
```

# GROUPING DATA USING GROUPBY

```python
fortune = pd.read_csv("fortune1000.csv")
fortune
```

```python
sectors = fortune.groupby("Sector")
sectors
```

```python
sectors.groups
```

```python
sectors.first()
```

```python
sectors.last()
```

```python
sectors.nth(0)      # first occurance
sectors.nth(1)      # second occurance
sectors.nth(2)      # third occurance
```

```python
sectors.get_group("Energy")
```

```python
sectors.sum()   # count(), max(), min(), median()
```

```python
sectors["Revenues"].sum()
```

```python
aggregations = {
    "Revenues":"min",
    "Profits":"max",
    "Employees":"mean"
}
sectors.agg(aggregations)
```

```python
sector_and_industry = fortune.groupby(["Sector", "Industry"])
sector_and_industry.get_group(("Business Services", "Education"))
```
